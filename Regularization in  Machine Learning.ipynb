```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer(strategy='mean')
```

# Encoding

```python
cars['make'].nunique()
```

```python
from sklearn.preprocessing import OrdinalEncoder   #only on categorical column not numerical column
```

```python
cars.head()
```

```python
categorical_columns = cars.select_dtypes(object).columns
```

```python
categorical_columns
```

```python
oe = OrdinalEncoder()
```

```python
cars[categorical_columns] = oe.fit_transform(cars[categorical_columns])
```

```python
cars.head()
```

# Model Building

```python
from sklearn.model_selection import train_test_split
```

```python
X = cars.iloc[:,:-1]
Y = cars.iloc[:,-1]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25,random_state=1)
```

Linear Regression

```python
from sklearn.linear_model import LinearRegression,Ridge , Lasso
```

```python
lr = LinearRegression()

lr.fit(X_train,Y_train)
ylr = lr.predict(X_test)

print(lr.score(X_train,Y_train))
print(lr.score(X_test,Y_test))
```

```python
lr.coef_
```

```python
from sklearn.metrics import mean_squared_error , r2_score
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mse = mean_squared_error(Y_test,ylr)
mae = mean_absolute_error(Y_test,ylr)
rmse = np.sqrt(mse)

print('MAE: ',mae)
print('RMSE: ',rmse)
print('The mean squares value is :',mse)
print('The R2 score is:',r2_score(Y_test,ylr))
```

# Lasso Regression

```python
lasso = Lasso()
lasso.fit(X_train,Y_train)

lassoy = lasso.predict(X_test)
print(lasso.score(X_train,Y_train))
print(lasso.score(X_test,Y_test))
```

```python
lasso.coef_
```

```python
for i in range(10,100):
    lasso = Lasso(alpha=i)
    lasso.fit(X_train,Y_train)
    
    lassoy = lasso.predict(X_test)
    trainac = lasso.score(X_train,Y_train)
    testac=lasso.score(X_test,Y_test)
    print(f" {i} Train {trainac} Test {testac} ")
```

```python
lasso = Lasso(alpha=99)
lasso.fit(X_train,Y_train)

lassoy = lasso.predict(X_test)
print(lasso.score(X_train,Y_train))
print(lasso.score(X_test,Y_test))
```

# Ridge regression

```python
ridge = Ridge()
ridge.fit(X_train,Y_train)
y_ridge = ridge.predict(X_test)
print(ridge.score(X_train,Y_train))
print(ridge.score(X_test,Y_test))
```
